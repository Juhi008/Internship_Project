## Project Title :  Marketing Campaign for Banking Products.





The bank which has a growing customer base. Majority of these customers are liability customers (depositors) with varying size of deposits. The number of customers who are also borrowers (asset customers) is quite small.So,the management wants to explore ways of converting its liability customers to personal loan customers. A campaign that the bank ran last year for liability customers showed a healthy conversion rate of over 9% success. This has encouraged the retail marketing department to devise campaigns to better target marketing to increase the success ratio with a minimal budget.

The department wants to build a model that will help them identify the potential customers who have a higher probability of purchasing the loan. This will increase the success ratio while at the same time reduce the cost of the campaign.


The dataset contains data on 5000 Customers and among these only 480 (=9.6%) accepted the personal loan that was offered to them in the earlier campaign.


##      

## Objective:
     
   To predict the likelihood of a liability Customer buying Personal Loan.

## Attribute Information:

● ID: Customer ID

● Age: Customer's age in completed years

● Experience: #years of professional experience

● Income: Annual income of the customer ($000)

● ZIP Code: Home Address ZIP code.

● Family: Family size of the customer

● CCAvg: Avg. spending on credit cards per month ($000)

● Education: Education Level. 1: Undergrad; 2: Graduate; 3:Advanced/Professional

● Mortgage: Value of house mortgage if any. ($000)

● Personal Loan: Did this customer accept the personal loan offered in the last
  campaign?

● Securities Account: Does the customer have a securities account with the bank?

● CD Account: Does the customer have a certificate of deposit (CD) account with
  the bank?
  
● Online: Does the customer use internet banking facilities?

● Credit card: Does the customer use a credit card issued by the bank?

##  4. Transformation , Normalisation of  data and split the data into training and test set in                  the ratio of 70:30


```python
# importing required library.

from sklearn.preprocessing import PowerTransformer as pt
```

```python
# put all the independent values in variable x and dependent values in a variable y.

x = df.drop(['Personal Loan'], axis=1)
y = df['Personal Loan']
```

```python
pt1 = pt(method='yeo-johnson',standardize=False)
pt1.fit(x['Income'].values.reshape(-1,1))
temp = pt1.transform(x['Income'].values.reshape(-1,1))

sns.distplot(x['Income'])
plt.show()
```

```python
pt1.fit(x["CCAvg"].values.reshape(-1,1))
temp = pt1.transform(x["CCAvg"].values.reshape(-1,1))
sns.distplot(temp)
```

After defining the Independent and Dependent variable, Split the data into training and test set.

```python
#splitting data into Training and test set.

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3, random_state=1)

```

```python
#checking dimension of given set.
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Feature Scaling library
from sklearn.preprocessing import StandardScaler as ss
ss1=ss()
x_train=ss1.fit_transform(x_train)
x_test=ss1.fit_transform(x_test)
x_train,x_test
```

## 5. Logistic Regression Model

     Now we are going to use Logistic Regression Model to predict the likelihood of a customer buying Personal loan

Logistic Regression is a Supervised Learning Classification Algorithms, used to predict the probability of a target variable.

It is used to assign obsevation to a Discrete set of classes.

```python
# importing library

from sklearn.linear_model import LogisticRegression
```

```python
# creating instance of logistic regression.

log= LogisticRegression()
```

```python
# fitting the model.
log.fit(x_train, y_train)
```

## Bagging Ensemble Classifier Model

In ensemble algorithms, bagging methods form a class of algorithms which build several instances of a black-box estimator on random subsets of the original training set and then aggregate their individual predictions to form a final prediction. These methods are used as a way to reduce the variance of a base estimator (e.g., a decision tree), by introducing randomization into its construction procedure and then making an ensemble out of it. In many cases, bagging methods constitute a very simple way to improve with respect to a single model, without making it necessary to adapt the underlying base algorithm. As they provide a way to reduce overfitting, bagging methods work best with strong and complex models (e.g., fully developed decision trees), in contrast with boosting methods which usually work best with weak models (e.g., shallow decision trees).
